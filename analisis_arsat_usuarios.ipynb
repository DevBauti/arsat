```python
#lowercase
refefo_nodes.columns = refefo_nodes.columns.str.lower()
refefo_nodes['provincia'] = refefo_nodes['provincia'].str.lower()
refefo_nodes['localidad'] = refefo_nodes['localidad'].str.lower()
refefo_nodes['plan'] = refefo_nodes['plan'].str.lower()
#float and round
refefo_nodes['latitud'] = refefo_nodes['latitud'].str.replace(',', '.').astype(float).round(5)
refefo_nodes['longitud'] = refefo_nodes['longitud'].str.replace(',', '.').astype(float).round(5)
#int
refefo_nodes['poblacion'] = refefo_nodes['poblacion'].astype(int)

refefo_nodes.info()
```

```python
#function draw graphic

def draw_bar_sorted(column, df):

  count = df[column].value_counts().reset_index()
  count.columns = [column, 'frequency']
  count = count.sort_values(by='frequency', ascending=True)

  fig = px.bar(count, x=column, y='frequency',
               title=f'Frequency of {column} (sorted)')
  fig.update_layout(xaxis={'categoryorder':'total ascending'})
  fig.show()
```

```python
draw_bar_sorted('provincia', refefo_nodes)
```

```python
def draw_bar_sorted_2(column, df):
  fig = px.bar(df, x='provincia', y=column,
               title=f'Frequency of {column} (sorted)')
  fig.update_layout(xaxis={'categoryorder':'total ascending'})
  fig.show()
```

```python
draw_bar_sorted_2('clientes', refefo_clients)
```

```python
#maps
def map(df, lat, lon):
  fig = px.scatter_mapbox(df, lat=lat, lon=lon,
                          color_discrete_sequence=['red'],
                          zoom=4, height=800)
  fig.update_layout(mapbox_style='carto-darkmatter')
  fig.update_layout(margin={'r':0, 't':0, 'l':0, 'b':0}, )
  fig.show()
```

```python
# copy
df = refefo_nodes.copy()
#draw
map(df, 'latitud', 'longitud')
```
